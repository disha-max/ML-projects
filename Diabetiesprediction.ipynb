Importing the libraries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

data collection and analysis

pima dataset

```python
diabetes_data= pd.read_csv('/content/diabetes.csv')
```

```python
diabetes_data.head()
```

```python
diabetes_data.shape
```

```python
diabetes_data.describe()
```

```python
diabetes_data['Outcome'].value_counts()
```

0 represents nondiabetic
1 represents diabetic


```python
diabetes_data.groupby('Outcome').mean()
```

```python
x=diabetes_data.drop(columns='Outcome', axis=1)
y=diabetes_data['Outcome']
print(x)
print(y)
```

Data Standerdization

```python
scaler=StandardScaler()
```

```python
scaler.fit(x)
```

```python
standardized_data=scaler.transform(x)
```

```python
print(standardized_data)
```

```python
x=standardized_data
print(x)
```

```python
print(y)
```

Train and test data

```python
x_test, x_train, y_test, y_train=train_test_split(x,y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
classifier=svm.SVC(kernel='linear')
```

```python
classifier.fit(x_train, y_train)
```

```python
#model evaluation
x_train_prediction=classifier.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction, y_train)
```

```python
print(training_data_accuracy)
```

```python
#model evaluation
x_test_prediction=classifier.predict(x_test)
testing_data_accuracy=accuracy_score(x_test_prediction, y_test)
print(testing_data_accuracy)
```

Making a predictive System

```python
input_data= (7,	83,	65,	39,	0,	27.6,	0.371, 33)


input_data_as_numpy_array=np.asarray(input_data)

input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)



std_data=scaler.transform(input_data_reshaped)
print(std_data)

prediction=classifier.predict(std_data)
print(prediction)
if(prediction[0]==1):
  print("paitent has diabetes")
elif(prediction[0]==0):
  print("nondiabetic")
else:
  print("falied prediction")
```
